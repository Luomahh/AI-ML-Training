```python
score = eval(input("Enter in your score:"))

if score >= 70 and score <= 100:
    print("A")
elif score >= 60 and score < 70:
    print("B")
elif score >= 50 and score < 60:
    print("C")
elif score >= 0 and score < 50:
    print("F")
else:
    print("Invalid number")
```
